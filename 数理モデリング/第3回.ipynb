演習6：ロジスティック写像のコンピュータ実験（3）

```python

import numpy as np
import matplotlib.pyplot as plt
mu =3.8282
N=10
x=np.zeros(N+1)
y=np.zeros(N+1)
print(x,y)
x[0]=0.1
y[0]=0
for i in range(N):
  x[i+1]=mu*x[i]*(1-x[i])
  if x[i]<=1/2:
    y[i]=0
  else:
    y[i]=1
print(x,y)
plt.plot(x,'-o')
plt.plot(y,'-k')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

n=10
x=np.zeros(n+1)
x[0]=0.0
dx=1.0
for i in range(n):
  a=random.random()
  if a<0.5:
    x[i+1]=x[i]+dx
  else:
    x[i+1]=x[i]-dx
plt.plot(x,'.-')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

mu=4.0
N=100
x=np.zeros(N+1)
y=np.zeros(N+1)
x[0]=0.1
y[0]=x[0]+pow(10,-10)
for i in range(N):
  x[i+1]=mu*x[i]*(1.0-x[i])
  y[i+1]=mu*y[i]*(1.0-y[i])
plt.plot(x,'-+')
plt.plot(y,'-y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu=4.0
N=100
x=np.zeros(N+1)
y=np.zeros(N+1)
x[0]=0.1
y[0]=x[0]+pow(10,-10)
for i in range(N):
  x[i+1]=mu*x[i]*(1.0-x[i])
  y[i+1]=mu*y[i]*(1.0-y[i])
plt.plot(np.abs(y-x),'-+')
plt.show()
```
